```python
import matplotlib.pyplot as plt

# Data for past three years
years = ["2020/21", "2021/22", "2022/23"]
imports = [200, 1807, 4050]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(years, imports, marker='o', linestyle='-', color='blue')
plt.xlabel("Fiscal Year")
plt.ylabel("Number of Imported EVs")
plt.title("Growth of Electric Vehicle Imports in Nepal")

# Add annotations for percentage increase
for i, year in enumerate(years):
    if i > 0:
        percentage_increase = (imports[i] - imports[i-1]) / imports[i-1] * 100
        plt.annotate(f"{percentage_increase:.2f}%\nincrease", (years[i], imports[i]), textcoords='offset points', xytext=(-10, 10))

plt.grid(True)
plt.show()
```

```python
# Define data
years = ["2016/17", "2020/21"]
incentives = ["Public", "Private"]
duty_rates = [[1, 10], [40, 40]]  # Customs duty rates for each year and incentive

# Create bar plot
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(years))

for i, incentive in enumerate(incentives):
    plt.bar(index[i::2], duty_rates[i], bar_width, label=incentive, color=["green", "red"][i])

plt.xlabel("Year")
plt.ylabel("Customs Duty (%)")
plt.xticks([x + bar_width/2 for x in index], years)
plt.title("Customs Duty for Public and Private EVs in Nepal")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Define data
power_levels = ["Low Power", "High Power"]
duty_rates = [30, 80]  # Excise duty rates for each power level

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(duty_rates, labels=power_levels, autopct="%1.1f%%", startangle=90, colors=["orange", "gold"])
plt.title("Excise Duty for Different Motor Power Levels in 2020/21")
plt.axis("equal")  # Equal aspect ratio for a circular chart
plt.tight_layout()
plt.show()
```
