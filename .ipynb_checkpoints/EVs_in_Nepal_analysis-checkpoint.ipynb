```python
import matplotlib.pyplot as plt

# Data for past three years
years = ["2020/21", "2021/22", "2022/23"]
imports = [200, 1807, 4050]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(years, imports, marker='o', linestyle='-', color='blue')
plt.xlabel("Fiscal Year")
plt.ylabel("Number of Imported EVs")
plt.title("Growth of Electric Vehicle Imports in Nepal")

# Add annotations for percentage increase
for i, year in enumerate(years):
    if i > 0:
        percentage_increase = (imports[i] - imports[i-1]) / imports[i-1] * 100
        plt.annotate(f"{percentage_increase:.2f}%\nincrease", (years[i], imports[i]), textcoords='offset points', xytext=(-10, 10))

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define data
years = ["2016/17", "2020/21"]
incentives = ["Public", "Private"]
duty_rates = [[1, 10], [40, 40]]  # Customs duty rates for each year and incentive

# Create bar plot
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(years))

for i, incentive in enumerate(incentives):
    plt.bar(index[i::2], duty_rates[i], bar_width, label=incentive, color=["green", "red"][i])

plt.xlabel("Year")
plt.ylabel("Customs Duty (%)")
plt.xticks([x + bar_width/2 for x in index], years)
plt.title("Customs Duty for Public and Private EVs in Nepal")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
from geopy.geocoders import Nominatim

# Define all locations and districts
locations = [
    ("Dharke", "Dhading"),
    ("Mugling", "Chitwan"),
    ("Sajha Sewa Balkhu", "Kathmandu"),
    ("Sajha Sewa Pulchowk", "Lalitpur"),
    ("Nepal Bidhut Pradhikaran, Ratnapark", "Kathmandu"),
    ("Nepal Police Club, Bhrikutimandap", "Kathmandu"),
    ("Paras Buspark", "Chitwan"),
    ("Khurkot", "Sindhuli"),
    ("Sindhuli Buspark", "Sindhuli"),
    ("Ilaka Prahari Karyalaya, Bardibas", "Mahottari"),
    # ... and add all remaining locations ...
    ("Lamki", "Kailali"),
]

# Get coordinates using Nominatim
geolocator = Nominatim(user_agent="my_app")
coordinates = []
for location, district in locations:
    try:
        coordinates.append(geolocator.geocode(f"{location}, {district}"))
    except:
        print(f"Error geocoding {location}, {district}")
        continue

# Create a map with Plotly
fig = px.scatter_mapbox(
    lat=[coord.latitude for coord in coordinates],
    lon=[coord.longitude for coord in coordinates],
    hover_data={"location": [f"{location} ({district})" for location, district in locations]},
    color="green",
    zoom=6,
    mapbox_style="stamen-terrain",
)

# Configure map and show it
fig.update_layout(
    title="Charging Stations in Nepal",
    margin=dict(l=0, r=0, b=0, t=0),
    legend=dict(title="Charging stations"),
    mapbox_sliders=dict(
        currentlayer_thickness=5,
        baseslidecolor="#f5f5f5",
        activecolor="#333333",
        show=True,
        location="bottom",
    ),
)
fig.show()
```
